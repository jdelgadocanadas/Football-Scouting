# Eliminar los % de determinadas columnas para transformarlos en 'int' y poder trabajar con ellos

```python
def no_percents (player_name,name_column):
     
     
    a = df.loc [player_name, name_column]
    
    if type (a) == str:
        b = a [:-1]
        result = int (b)
    else:
        result = a
      
    return result

no_percents ('Rayo Vallecano','Chances, % of conversion')
```

# Partidos jugados

```python
# Antes de normalizar los datos, debemos dividir las estad'isticas por los partidos jugados
# 1.Creo una fx que introduciendo nombre jugadora me devuelve los partidos jugados

def partidos_jugados (player_name):
    
    result = df.loc [player_name, 'Number of games']
    return result

partidos_jugados ('Rayo Vallecano')
```

# Funcion Statsteam

```python
# Junto todos los pasos anteriores para crear la fx 'Stats_player'
# En ella, introducimos el nombre de una jugadora y nos devuelve sus estadisticas/ partido

def Stats_team (name_team):
    
    grouped = liga.groupby('Team')
    a = grouped.get_group(name_team)
    a = a / partidos_jugados (name_team)
    a ['Matches played'] = [partidos_jugados (name_team)]
    a ['Chances, % of conversion'] = [no_percents ((name_team),'Chances, % of conversion')]
    a ['Accurate crosses, %'] = [no_percents ((name_team),'Accurate crosses, %')]
    a ['Challenges won, %'] = [no_percents ((name_team),'Challenges won, %')]
    a ['Challenges in defence won, %'] = [no_percents ((name_team),'Challenges in defence won, %')]
    a ['Challenges in attack won, %'] = [no_percents ((name_team),'Challenges in attack won, %')]
    a ['Air challenges won, %'] = [no_percents ((name_team),'Air challenges won, %')]
    a ['Successful dribbles, %'] = [no_percents ((name_team),'Successful dribbles, %')]
    a ['Tackles won, %'] = [no_percents ((name_team),'Tackles won, %')]
    a ['Accurate passes, %'] = [no_percents ((name_team),'Accurate passes, %')]
    a ['Pressing efficiency, %'] = [no_percents ((name_team),'Pressing efficiency, %')]
    a ['Ball possession, %'] = [no_percents ((name_team),'Ball possession, %')]
 
 
    return a

Stats_team ('Real Betis')    
```

```python
equipos = []

for k in equipos_liga:
    if k not in equipos:
        equipos.append (k)
equipos = equipos [1:]

equipos   
```

```python
def stats_season (equipos):
    
    tabla = Stats_team ('Eibar')  
    
    for k in (equipos):
        
        if k != 'Eibar':
            grouped = liga.groupby('Team')
            a = grouped.get_group(k)
            a = a / partidos_jugados (k)
            a ['Matches played'] = [partidos_jugados (k)]
            a ['Chances, % of conversion'] = [no_percents ((k),'Chances, % of conversion')]
            a ['Accurate crosses, %'] = [no_percents ((k),'Accurate crosses, %')]
            a ['Challenges won, %'] = [no_percents ((k),'Challenges won, %')]
            a ['Challenges in defence won, %'] = [no_percents ((k),'Challenges in defence won, %')]
            a ['Challenges in attack won, %'] = [no_percents ((k),'Challenges in attack won, %')]
            a ['Air challenges won, %'] = [no_percents ((k),'Air challenges won, %')]
            a ['Successful dribbles, %'] = [no_percents ((k),'Successful dribbles, %')]
            a ['Tackles won, %'] = [no_percents ((k),'Tackles won, %')]
            a ['Accurate passes, %'] = [no_percents ((k),'Accurate passes, %')]
            a ['Pressing efficiency, %'] = [no_percents ((k),'Pressing efficiency, %')]
            a ['Ball possession, %'] = [no_percents ((k),'Ball possession, %')]
           
            
    

        tabla = tabla.append(a, ignore_index = False)
        
    
    return tabla
```

```python
tabla = stats_season (equipos)    
tabla                                 
```
